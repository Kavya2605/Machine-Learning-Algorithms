DataFrame vs Series

A DataFrame is a 2-dimensional data structure designed for handling tabular data with multiple columns, while a Series is a 1-dimensional data structure used to represent a single column or row of data within a DataFrame or as a standalone data structure.

```python
#  Load the Titanic Dataset
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv")
```

**1. survived (int)**

0 → Passenger did not survive
1 → Passenger survived

Target variable in predictive modeling.

**2. pclass (int)**
Passenger class (proxy for socio-economic status):

1 → Upper class

2 → Middle class

3 → Lower class

**3. sex (object/string) **

Passenger gender:

male

female

**4. age (float)**
Age in years (some values are missing).

Children typically have lower ages; missing ages often need imputation.

**5. sibsp (int)**
Number of siblings or spouses aboard:

0 → Traveling alone

>0 → Traveling with family (brothers, sisters, husband, wife)

6. parch (int)
Number of parents or children aboard:

0 → No parents/children with them

>0 → With family (mother, father, son, daughter)

7. fare (float)
Ticket fare paid by passenger (in British Pounds).

8. embarked (category/string)
Port of embarkation (where passenger boarded):

C → Cherbourg

Q → Queenstown

S → Southampton

9. class (category)
Passenger class as string label:

First, Second, Third

Derived from pclass.

10. who (category)
A simplified gender/age categorization:

man → adult male

woman → adult female

child → age < 16

11. adult_male (bool)
True → Passenger is an adult male

False → otherwise (woman or child)

12. deck (category)
Cabin deck letter (A–G, or NaN if unknown).

Gives approximate location of cabin.

13. embark_town (category/string)
Full name of embarkation port:

Cherbourg, Queenstown, Southampton

Derived from embarked.

14. alive (category)
String version of survival status:

yes → survived

no → did not survive

Derived from survived.

15. alone (bool)
Indicates if passenger was traveling alone:

True → no siblings/spouses & no parents/children (i.e., sibsp + parch == 0)

False → traveling with family.

df.describe() generates summary statistics of numeric columns in a DataFrame by default.
It’s one of the most commonly used methods in Exploratory Data Analysis (EDA).

1. For numeric columns:

count → number of non-null values

mean → average value

std → standard deviation

min → minimum value

25% → first quartile (Q1)

50% → median (Q2)

75% → third quartile (Q3)

max → maximum value

2. For categorical columns (include='all'):

count → number of non-null values

unique → number of unique values

top → most frequent value

freq → frequency of the top value




```python
df.describe()
```

```python
df.describe(include='all')
```

**df.isnull().sum()**

1st part: df.isnull()

Returns a DataFrame of the same shape as df.

Each cell contains:

True → if the value is NaN (missing)

False → if the value is not missing

2nd part: .sum()

When applied column-wise (default), it counts True values (since True = 1, False = 0).

The result is a Series showing how many missing values each column has.

```python
df.isnull().sum()
```

##  Section A: Filtering and Conditional Selection

- Selecting specific columns  
- Applying conditions with boolean indexing  
- Using `.loc[]` and `.iloc[]`


```python
# Selecting Columns
df[['sex', 'age', 'fare']].head()
```

```python
# Filtering: Age greater than 50
df[df['ticket_fare'] > 30]
```

```python
# Females who paid fare > 50
df[(df['sex'] == 'female') & (df['fare'] > 50)].head()
```

```python
# Embarked in C or Q
df[df['embarked'].isin(['C'])]
```

```python
df['embarked'].value_counts()
```

unique()
Shows all unique values in a column.

```python
df['embarked'].unique()
```

nunique()
Counts number of unique values.

```python
df['embarked'].nunique()
```

pd.crosstab() (Cross-tabulation)
Compares two categorical columns (like a pivot table).

```python
pd.crosstab(df['sex'], df['embarked'])
```

```python
# Not male
df[~df['sex'].isin(['male'])]
```

 `loc[]` is label-based indexing, `iloc[]` is position-based indexing.

```python
# Using .loc to select specific rows and columns
df.loc[df['age'] > 60, ['sex', 'age']]
```

```python
# loc[] — label-based
df.loc[0:10, ['sex', 'age']]
```

```python
# iloc[] — integer position based (iloc does not care about column names, only positions)
df.iloc[0:5, [2, 4, 9]]  # Same columns but by position
```

##  Section B: Grouping and Aggregations

We use `.groupby()` to group data based on columns, then apply aggregation functions.


```python
# Average fare by gender
df.groupby('sex')['fare'].mean()
```

```python
# Average age by gender and class
df.groupby(['sex', 'class'])['age'].mean()
```

```python
# Multiple aggregations
df.groupby('class').agg({
    'fare': ['mean', 'max', 'min'],
    'age': 'median'
})
```

```python
df['fare_mean_by_class'] = df.groupby('class')['fare'].transform('mean')
df.head(20)

```

```python
df.groupby('class')['fare'].agg('mean')
```

## Section E: Cleaning and Saving Data

Common cleaning steps:
- Fill or drop missing values
- Rename columns
- Save cleaned data


```python
# Fill missing ages with mean age
df['age'] = df['age'].fillna(df['age'].mean())
df.head()
```

```python
df['deck'].isnull().sum()
```

```python
# Drop rows where 'embarked' is missing
df.dropna(subset=['embarked'], inplace=True)
df
```

```python
# Rename column
df.rename(columns={'fare': 'ticket_fare'}, inplace=True)
df
```

```python
# Save to CSV
df.to_csv('titanic_clean.csv', index=False)
```
